```python
def categorize_garasi(jumlah_garasi):
    if jumlah_garasi <= 2:
        return 'Kecil'
    elif jumlah_garasi <= 5:
        return 'Sedang'
    else:
        return 'Besar'

# Buat kolom kategori baru berdasarkan jumlah garasi
df['Kategori Garasi'] = df['Garasi'].apply(categorize_garasi)
df.drop('Garasi', axis=1, inplace=True)
```
